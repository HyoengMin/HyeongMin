```python
import requests
import json

response = requests.get('https://news.daum.net/api/harmonydic/contents/news.json?category=economic&approved=true&page=1&pageSize=20&pagesToShow=10&range=1')

data = json.loads(response.text)
news=data['list']
```

```python
# 20개 뉴스기사의
# 날짜, 제목, 링크, 언론사

for new in data['list']:
    date=new['newsId'][:8]
    title=new['title']
    link=new['contentUrl']
    press=new['cpKorName']

    print(date,title,link,press)
```

```python

for i in range(20):
    news_data=news[i]['newsId'][:8]
    news_title=news[i]['title']
    news_link=news[i]['contentUrl']
    news_press=news[i]['cpKorName']
    
    
    
    print(news_data,news_title,news_link,news_press)
```
